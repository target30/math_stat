```python
import numpy as np
import matplotlib
import skimage.io
import matplotlib.pyplot as plt
import random
import math
import statistics 
from scipy.stats import skew
from scipy.stats import expon 
from sklearn.neighbors import KernelDensity
N=25
number_of_element=np.linspace(1, N, N)
rand_number=[random.uniform(0, 1) for i in range(N)]
my_distribution=[]
for i in range (N):
    my_distribution.append(round(-math.log(1-rand_number[i]),8))
distribution= expon.rvs(size = N)
print ("выборка с помощью готовых функций :\n", distribution)
print ("выборка генерируемая мной :\n", my_distribution)
```

# Сортируем выборки:

```python
my_distribution.sort()
distribution.sort()
print ("выборка с помощью готовых функций :\n", distribution)
print ("выборка генерируемая мной :\n", my_distribution)
```

```python
plt.plot(my_distribution,'blue' ,  distribution , 'r' )
```

# Работаем с выборкой сделанной с помощью готовых функций

# Пункт а)мода, медиана, размах, коэффициент асимметрии

```python
moda=statistics.mode(distribution)
mediana=statistics.median(distribution)
scope=max(distribution)-min(distribution)
skew_coefficient=skew(distribution, axis=0, bias=True)
print('мода:',moda )
print('медиана:',mediana)
print('размах:',scope)
print('коэффициент асимметрии:',skew_coefficient)
```

# Пункт Б)Построить эмпирическую функцию распределения,гистограмму и boxplot

```python
x = np.linspace(0, np.max(distribution) + 0.5, 1000)
y = np.array([np.sum(distribution<i) for i in x]) / len(distribution)
plt.scatter(x, y, color = 'red')
```

```python
k=int(1+math.log2(N))
plt.hist(distribution, color = 'blue', edgecolor = 'black',
         bins = k)
```

```python
matplotlib.pyplot.boxplot(distribution)
```

# В)Ядерная оценка

```python
kde = KernelDensity(kernel='epanechnikov', bandwidth=0.9).fit(distribution[np.newaxis, :].T)
X_plot = np.linspace(0, 10, 1000)[:, np.newaxis]
log_dens = kde.score_samples(X_plot)
plt.plot(X_plot[:, 0], np.exp(log_dens), color='blue',
        linestyle='-', label="kernel = epanechnikov")
```

# Г)Определить плотность распределения среднего арифметического элементов выборки. Сравнить с бутстраповской оценкой плотности

Мат.ожидание случайной величины=1, дисперсия=1.
Плотность распределения среднего арифметического эл-тов выборки = 1+N(0,1).
Проверим:

```python
average_values=[]
for i in range (1000):
    average_value=0
    positions_of_elements=[random.randint(0, N-1) for i in range(N)]
    for position in positions_of_elements:
        average_value+=distribution[position]
    average_value/=N
    average_values.append(average_value)
    
```

```python
plt.hist(average_values, color = 'blue', edgecolor = 'black',
         bins = int(1+math.log2(1000)))
```

# Д) Бутстраповская оценка плотности распределения коэффициентов асимметрии


```python
skew_coefficientes=[]
for i in range (1000):
    average_value=0
    positions_of_elements=[random.randint(0, N-1) for i in range(N)]
    samples=[]
    for position in positions_of_elements:
        samples.append(distribution[position])
    skew_coefficiente_of_new_samples=skew(samples, axis=0, bias=True)
    skew_coefficientes.append(skew_coefficiente_of_new_samples)
```

```python
plt.hist(skew_coefficientes, color = 'blue', edgecolor = 'black',
         bins = int(1+math.log2(1000)))
```

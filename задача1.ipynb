```python
import numpy as np
import matplotlib.pyplot as plt
import random as random_number
teta=50
N=100
x= list(range(0, N))
numbers = [random_number.randint(0, teta) for i in range(0,N)]
print(numbers)
```

```python
teta_calculate1=[0] * N
teta_calculate2=[0] * N
teta_calculate3=[0] * N
teta_calculate4=[0] * N
teta_calculate5=[0] * N
for i in range(1,N):
    teta_calculate1[i]=2*sum(numbers[0:i])/i
    teta_calculate2[i]=min(numbers[0:i])*(i+1)
    teta_calculate3[i]=max(numbers[0:i])*(i+1)/i
    teta_calculate4[i]=min(numbers[0:i])+max(numbers[0:i])
    teta_calculate5[i]=numbers[0]+sum(numbers[1:i])/(i+1)
fig, ax = plt.subplots(figsize=(10, 8))
ax.scatter(x, teta_calculate1,
           c = 'green',
           s = 3, label = "feger")
ax.scatter(x, teta_calculate2,
           c = 'red',
           s = 3, label = "2")
ax.scatter(x, teta_calculate3,
           c = 'blue',
           s = 3, label = "3")
ax.scatter(x, teta_calculate4,
           c = 'orange',
           s = 3, label = "2")
ax.scatter(x, teta_calculate5,
           c = 'yellow',
           s = 3, label = "3")
plt.legend(["среднее" , "минимум", "максимум","минимум+максимум", " x0+среднее"])
```
